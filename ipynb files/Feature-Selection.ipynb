```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

## for feature selection
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv('../Data/data-created/train_preprocessed.csv')
```

```python
dataset.head()
```

```python
y_train = dataset["SalePrice"]
```

```python
X_train = dataset.drop(['Id','SalePrice'],axis=1)
```

```python
##Applying Lasso regularization to perform feature selection
#irst, specifying the Lasso model and then selecting a suitable alpha value(penalty value)
#Alpha and number of features are inversely proportional

#Then using the selectFromModel to select features with non-zero beta coefficients 
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) #Imp to set seed, the same random state to be used with the testing dataset
feature_sel_model.fit(X_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
# Printing the total and selected features
# Making a list of selected features
selected_features = X_train.columns[feature_sel_model.get_support()]

print(f'Total number of features : {len(X_train.columns)}')
print(f'Total number of selected features : {len(selected_features)}')
print(f'Features with beta coefficients shrank to zero : {np.sum(feature_sel_model.estimator_.coef_ == 0)}')
```

```python
selected_features
```

```python
X_train = X_train[selected_features]
```
