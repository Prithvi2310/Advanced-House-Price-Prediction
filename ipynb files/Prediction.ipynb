```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
%matplotlib inline

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

pd.set_option('display.max_columns', None)
```

```python
test_data = pd.read_csv('../Data/data-created/test_preprocessed.csv')
```

```python
test_data
```

```python
y_test = test_data['SalePrice']
X_test = test_data.drop(['Id','SalePrice'],axis=1)
```

```python
selected_features = np.array(['MSSubClass', 'MSZoning', 'Neighborhood', 'OverallQual', 'OverallCond',
       'YearRemodAdd', 'BsmtQual', 'BsmtExposure', 'BsmtFinSF1', 'TotalBsmtSF',
       'HeatingQC', 'CentralAir', '1stFlrSF', 'GrLivArea', 'KitchenQual',
       'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageCars', 'GarageCond',
       'PavedDrive'], dtype = object)
```

```python
selected_features
```

```python
X_test = X_test[selected_features]
X_test
```

#### Ordinary Least Squares model
##### Linear Regression 

```python
# Loading the model
filepath = '../models/OLS_model.sav'
regressor = pickle.load(open(filepath, 'rb'))
```

```python
test_predict = regressor.predict(X_test)
```

```python
# Printing coefficients of the model to cross check
print(f'Coefficients : {regressor.coef_}')
```

```python
# model evaluation
print(f'Mean Squared Error : {np.round(mean_squared_error(y_test,test_predict),5)}')
print(f'The coefficient of determination : {np.round(r2_score(y_test, test_predict),5)}')
```
